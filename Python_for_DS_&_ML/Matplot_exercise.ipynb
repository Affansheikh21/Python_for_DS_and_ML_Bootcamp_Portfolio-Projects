```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
ax = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
ax.plot(x, y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```
